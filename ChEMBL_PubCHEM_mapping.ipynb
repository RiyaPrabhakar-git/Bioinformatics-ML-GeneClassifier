```python
import pandas as pd
import pubchempy as pcp
from tqdm import tqdm

csvfile = "/content/DOWNLOAD-EVmOHc5oXUtN9ioWHTacv6keFFE1XAtOTHmdNDAPsxU_eq_.csv"
# Load your ChEMBL CSV
df = pd.read_csv(csvfile, sep=';')

results = []

for idx, row in tqdm(df.iterrows(), total=len(df)):
    chembl_id = row["Molecule ChEMBL ID"]
    inchikey = row.get("standard_inchi_key", None)

    try:
        if pd.notna(inchikey):
            compounds = pcp.get_compounds(inchikey, 'inchikey')
        else:
            compounds = []

        if compounds:
            cid = compounds[0].cid
            results.append([chembl_id, inchikey, cid, "YES"])
        else:
            results.append([chembl_id, inchikey, None, "NO"])

    except Exception as e:
        results.append([chembl_id, inchikey, None, "ERROR"])

# Save mapping
out_df = pd.DataFrame(
    results,
    columns=["ChEMBL_ID", "InChIKey", "PubChem_CID", "Exists_in_PubChem"]
)

out_df.to_csv("chembl_pubchem_mapping.csv", index=False)
```

```python
import pubchempy as pcp
import os
from tqdm import tqdm
import pandas as pd # Import pandas here as well, if it's not guaranteed to be imported globally.

# Reload the original ChEMBL data since this process needs 'Molecule ChEMBL ID', 'Molecule Name', and 'Smiles'
df = pd.read_csv(csvfile, sep=';')

os.makedirs("/content/sdf_3d", exist_ok=True)

results = []

for i, row in tqdm(df.iterrows(), total=len(df)):
    chembl_id = row["Molecule ChEMBL ID"]
    name = row["Molecule Name"]
    smiles = row["Smiles"]

    print(f"\n Processing {chembl_id} | {name}")

    try:
        compounds = pcp.get_compounds(smiles, namespace="smiles")

        if not compounds:
            print(" Not found in PubChem")
            results.append([chembl_id, name, smiles, None, "NO"])
            continue

        compound = compounds[0]
        cid = compound.cid
        print(f"Found PubChem CID: {cid}")

        # Try downloading 3D structure
        compounds_3d = pcp.get_compounds(cid, "cid", record_type="3d")

        if compounds_3d:
            sdf = compounds_3d[0].to_sdf()
            file_path = f"/content/sdf_3d/{chembl_id}_CID{cid}.sdf"
            with open(file_path, "w") as f:
                f.write(sdf)

            print(f"3D SDF saved: {file_path}")
            results.append([chembl_id, name, smiles, cid, "YES"])
        else:
            print("No 3D structure available")
            results.append([chembl_id, name, smiles, cid, "2D_ONLY"])

    except Exception as e:
        print(" Error:", e)
        results.append([chembl_id, name, smiles, None, "ERROR"])
```

```python
out_df = pd.DataFrame(
    results,
    columns=["ChEMBL_ID", "Molecule_Name", "SMILES", "PubChem_CID", "Status"]
)

out_df.to_csv("chembl_pubchem_mapping.csv", index=False)
out_df
```
